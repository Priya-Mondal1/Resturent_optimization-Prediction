## Feature Engineering + Data Preprocessing.

#### Create Target Variables

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

from sklearn.impute import SimpleImputer

import warnings
warnings.filterwarnings("ignore")
```

```python
df.head()
```

```python
df["TotalOrders"] = (
    df["InStoreOrdersCount"] +
    df["UberEatsOrdersCount"] +
    df["DoorDashOrdersCount"] +
    df["SelfDeliveryOrdersCount"]
)
```

```python
df["TotalRevenue"] = (
    df["InStoreRevenue"] +
    df["UberEatsRevenue"] +
    df["DoorDashRevenue"] +
    df["SelfDeliveryRevenue"]
)
```

#### *Channel Revenue Ratios*

```python
df["InStoreRevenueRatio"] = (
    df["InStoreRevenue"] /
    df["TotalRevenue"]
)

df["UberRevenueRatio"] = (
    df["UberEatsRevenue"] /
    df["TotalRevenue"]
)

df["DoorDashRevenueRatio"] = (
    df["DoorDashRevenue"] /
    df["TotalRevenue"]
)

df["SelfDeliveryRevenueRatio"] = (
    df["SelfDeliveryRevenue"] /
    df["TotalRevenue"]
)
df["ProfitPerOrder"] = (
    df["TotalNetProfit"] /
    df["MonthlyOrders"]
)
```

### *Cost-to-Revenue Ratio*

```python
df["TotalCost"] = (
    df["TotalRevenue"] -
    df["TotalNetProfit"]
)

df["CostRevenueRatio"] = (
    df["TotalCost"] /
    df["TotalRevenue"]
)
```

### *Interaction Features*

#### Delivery Cost × Self Delivery Share

```python
df["Delivery_SD"] = (
    df["DeliveryCostOrder"] *
    df["SD_share"]
)
```

#### Commission × Uber Eats Share

```python
df["Commission_UE"] = (
    df["CommissionRate"] *
    df["UE_share"]
)
```

#### Growth Adjusted Orders

```python
df["GrowthAdjustedOrders"] = (
    df["MonthlyOrders"] *
    df["GrowthFactor"]
)
```

#### Average Revenue Per Order

```python
df["RevenuePerOrder"] = (
    df["TotalRevenue"] /
    df["MonthlyOrders"]
)
```

#### Margin %

```python
df["ProfitMargin"] = (
    df["TotalNetProfit"] /
    df["TotalRevenue"]
)
```

### *Missing Values*

```python
print(df.isnull().sum())
```

### *Remove Duplicates*

```python
df.drop_duplicates(inplace=True)
```

### *Outlier Handling (IQR)*

```python
numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:

    Q1 = df[col].quantile(0.25)

    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR

    upper = Q3 + 1.5 * IQR

    df[col] = np.where(
        df[col] < lower,
        lower,
        df[col]
    )

    df[col] = np.where(
        df[col] > upper,
        upper,
        df[col]
    )
```

### *Encode Categorical Columns*

```python
label = LabelEncoder()

cat_cols = [
    "RestaurantName",
    "CuisineType",
    "Segment",
    "Subregion"
]

for col in cat_cols:

    df[col] = label.fit_transform(df[col])
```

### *Select Features*

```python
X = df.drop(
    [
        "TotalNetProfit"
    ],
    axis=1
)

y = df["TotalNetProfit"]
```

### *Train-Test Split*

```python
X_train, X_test, y_train, y_test = train_test_split(

    X,
    y,

    test_size=0.20,

    random_state=42
)
```

Only needed for models like Linear Regression and XGBoost.

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

import numpy as np
import pandas as pd
```

# Profit Optimization Engine

### *Import Libraries*

```python
import numpy as np
import pandas as pd
```

### *Choose a Base Sample*

```python
base = X_test.iloc[[0]].copy()
```

### *Create Variables*

```python
best_profit = -999999999

best_configuration = None
```

#### Initially:

Best Profit = -999999999

Configuration = None

### *Optimization Loop*

```python
for instore in np.arange(0.30,0.60,0.05):

    for ue in np.arange(0.15,0.45,0.05):

        for dd in np.arange(0.10,0.30,0.05):

            sd = 1 - (ue + dd)

            if sd <= 0:
                continue

            for commission in np.arange(0.15,0.31,0.05):

                for radius in range(3,13):

                    sample = base.copy()

                    sample["InStoreShare"] = instore

                    sample["UE_share"] = ue

                    sample["DD_share"] = dd

                    sample["SD_share"] = sd

                    sample["CommissionRate"] = commission

                    sample["DeliveryRadiusKM"] = radius

                    profit = model.predict(sample)[0]

                    if profit > best_profit:

                        best_profit = profit

                        best_configuration = sample.copy()
```

```python
print("Best Predicted Profit")

print(best_profit)
```
